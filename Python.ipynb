## Static Methods
Static methods are used for faster code execution as we do not require to pass parameters like- self, cls. Also, they get initialized as soon as the class method is initialized

```python
import datetime

class car:
  base_price=100000 ## class variables
  def __init__(self,windows,doors,power):
    self.windows=windows
    self.doors=doors
    self.power=power
  def what_base_price(self):
    print('The base price is {}'.format(self.base_price))
  @classmethod
  def revise_base_price(cls,inflation):
    cls.base_price=cls.base_price+cls.base_price*inflation
  @staticmethod
  def check_year():
    if datetime.datetime.now().year==2026:
      return True
    else:
      return False
```

```python
car.check_year()
```

```python
car1= car(4,5,2000)
if(car1.check_year()):
  pass
else:
  car.revise_base_price()
```

```python
car1.base_price
```

# Multiple Inheritance

```python
class c(a):
  def method(self):
    print('C class method is called')
```

```python
class d(b,c):
  def method(self):
    print('D class method is called')
    c.method1(self)
    b.method1(self)
    a.method(self)
```
